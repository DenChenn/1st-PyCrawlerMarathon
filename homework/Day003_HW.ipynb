### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
#下載
from urllib.request import urlretrieve
path = "http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314"
urlretrieve(path,"./data/example.zip")

#解壓縮
import zipfile
unzip = './data/example.zip'
f = zipfile.ZipFile(unzip,"r")
f.extractall('./data')

#file I/O
IO = open("./data/64_72hr_CH.xml","r",encoding = "utf8")
fileIO = IO.read()
print (fileIO)
```

```python
#印出所有檔案名
import os

dirs = os.listdir("./data")

for filename in dirs:
    print(dirs)
```

```python
#xmltodict
import xmltodict
with open('./data/64_72hr_CH.xml',"r",encoding = 'utf8') as fd:
    doc = dict(xmltodict.parse(fd.read()))

print(doc['cwbopendata']['dataset']['datasetInfo']['datasetDescription'])
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
amount = doc['cwbopendata']['dataset']['locations']['location']
print(len(amount))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for ele in amount:
    print(ele['locationName'])
    print(ele['weatherElement'][0]['time'][0]['dataTime'])
    print(ele['weatherElement'][0]['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

print(amount[0]['locationName'])
timeline = amount[0]['weatherElement'][0]['time']
for ele in timeline:
    print(ele['dataTime'])
    print(ele['elementValue']['value'])
```
