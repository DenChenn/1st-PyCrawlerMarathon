## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
import json
data = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = json.loads(data.text)
print(len(response))
for imf in response:
    print(imf)
    print()
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for imf in response:
    print(imf['title'])
    print(imf['createdAt'])
    print(imf['commentCount'])
    print(imf['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
sumOfCp = 0
sumOfLk = 0
num = 0
for imf in response:
    sumOfCp = imf['commentCount'] + sumOfCp
    sumOfLk = imf['likeCount'] + sumOfLk
    num = num + 1

print(sumOfCp / num)
print(sumOfLk / num)
```
