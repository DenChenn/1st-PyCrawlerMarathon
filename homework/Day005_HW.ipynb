```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json
data = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = json.loads(data.text)
print(len(response))
for imf in response:
    print(imf)
    print()
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for imf in response:
    print(imf['title'])
    print(imf['createdAt'])
    print(imf['commentCount'])
    print(imf['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
sumOfCp = 0
sumOfLk = 0
num = 0
for imf in response:
    sumOfCp = imf['commentCount'] + sumOfCp
    sumOfLk = imf['likeCount'] + sumOfLk
    num = num + 1

print(sumOfCp / num)
print(sumOfLk / num)
```
